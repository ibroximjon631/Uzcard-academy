## 1. `successful(txs)`

**Задача:** отфильтровать список и вернуть только успешные транзакции (те, у которых `status == "ok"`).

**Зачем:** дальше многие метрики считаем только по успешным транзакциям - удобно один раз отфильтровать.

**Что делает:** возвращает новый список тех же словарей, без failed-транзакций. Порядок сохраняется.

- Вход: `txs` (`list[dict]`).
- Выход: `list[dict]`.

**Подсказка:** `[t for t in txs if t["status"] == "ok"]` или цикл с `if`.

```python
def successful(txs):
    return [t for t in txs if t["status"] == "ok"]
```

## 2. `total_volume(txs)`

**Задача:** посчитать общий оборот - сумму поля `amount` по переданным транзакциям.

**Зачем:** дневной оборот - базовая бизнес-метрика. Передавать сюда будем уже отфильтрованный список из `successful()`.

**Что делает:** просто складывает `amount` всех элементов входного списка и возвращает число.

- Вход: `txs` (`list[dict]`) - любой набор транзакций.
- Выход: `int` - сумма `amount`.

**Подсказка:** `sum(t["amount"] for t in txs)`.

```python
def total_volume(txs):
    return sum(t["amount"] for t in txs)
```

## 3. `top_gateway_by_volume(txs)`

**Задача:** найти платёжный шлюз, через который прошла самая большая сумма успешных транзакций.

**Зачем:** понять, какой шлюз приносит банку больше всего денег - аргумент для переговоров об эквайер-комиссии.

**Что делает:** среди **успешных** транзакций группирует `amount` по `gateway`, находит шлюз с максимальной суммой, возвращает его имя.

- Вход: `txs` (`list[dict]`) - **все** транзакции, включая failed (отбросьте failed сами).
- Выход: `str` - значение поля `gateway`, например `"UzCard"`.

**Подсказка:** заведите `dict` вида `{gateway: total_amount}`, накопите сумму в цикле, потом `max(d, key=d.get)`.

```python
def top_gateway_by_volume(txs):
    success_txs = [t for t in txs if t["status"] == "ok"]
    if not success_txs:
        return ""
    gateway_volumes = {}
    for t in success_txs:
        gw = t["gateway"]
        gateway_volumes[gw] = gateway_volumes.get(gw, 0) + t["amount"]
    return max(gateway_volumes, key=gateway_volumes.get)
```

## 4. `unique_merchants(txs)`

**Задача:** посчитать сколько разных торговых точек (`merchant_id`) встречается в данных.

**Зачем:** сколько разных мерчантов обслуживала система за день - индикатор покрытия сети.

**Что делает:** возвращает количество **уникальных** значений `merchant_id` по **всем** транзакциям, включая failed (факт обращения мерчанта засчитывается).

- Вход: `txs` (`list[dict]`).
- Выход: `int`.

**Подсказка:** `set` сам убирает дубликаты, `len(set(...))` даёт количество уникальных.

```python
def unique_merchants(txs):
    return len(set(t["merchant_id"] for t in txs))
```

## 5. `top_merchant_by_count(txs)`

**Задача:** найти самого активного мерчанта - того, у кого больше всего транзакций.

**Зачем:** самый активный мерчант - VIP-клиент; на него стоит обращать внимание (отдельный менеджер, индивидуальные тарифы).

**Что делает:** находит `merchant_id` с максимальным числом транзакций среди **всех** записей (включая failed - даже неуспешные попытки показывают активность).

- Вход: `txs` (`list[dict]`).
- Выход: `str` - `merchant_id`.

**Подсказка:** аналогично задаче 3 - словарь `{merchant_id: count}`, потом `max(...)`. Можно через `collections.Counter`, но и обычным циклом ок.

```python
def top_merchant_by_count(txs):
    if not txs:
        return ""
    merchant_counts = {}
    for t in txs:
        m_id = t["merchant_id"]
        merchant_counts[m_id] = merchant_counts.get(m_id, 0) + 1
    return max(merchant_counts, key=merchant_counts.get)
```

## 6. `avg_amount_in_age_range(txs, lo, hi)`

**Задача:** посчитать средний чек для клиентов определённой возрастной группы.

**Зачем:** средний чек по возрастной когорте - вход в маркетинговую сегментацию (молодые тратят меньше, средний возраст - крупные покупки).

**Что делает:** возьмите **успешные** транзакции, у которых `lo <= customer_age <= hi`, посчитайте средний `amount`, округлите до целого. Если ни одна транзакция не попала в диапазон - верните `0` (а не `ZeroDivisionError`).

- Вход: `txs` (`list[dict]`); `lo`, `hi` (`int`) - границы возраста, включительно.
- Выход: `int` (округлённое среднее) или `0`.

**Подсказка:** соберите `amount`-ы в список через list comprehension с двумя условиями (`status == "ok"` и проверка возраста), потом `round(sum(...) / len(...))`. Не забудьте проверку на пустоту.

```python
def avg_amount_in_age_range(txs, lo, hi):
    matched_amounts = [
        t["amount"] for t in txs
        if t["status"] == "ok" and lo <= t["customer_age"] <= hi
    ]
    if not matched_amounts:
        return 0
    return round(sum(matched_amounts) / len(matched_amounts))
```
